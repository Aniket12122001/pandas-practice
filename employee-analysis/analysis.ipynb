🎯 Objective

The objective of this project is to analyze employee data to extract meaningful insights for better workforce understanding and decision-making.

This analysis aims to:

    : How many total employees are there?

    : What is the average salary of employees?

    : Which employee has the highest salary?

    : What is the average salary for each department?

    : How many employees are there in each department?

    : Which city has the highest number of employees?

    : Find all employees with salary greater than 60000

    : Find employees with experience more than 4 years

    : Add a new column "Salary_Level" (High, Medium, Low based on salary)

    : Get top 3 highest paid employees

The goal is to simulate real-world HR analytics and support data-driven decision-making using data analysis techniques.



```python
import pandas as pd
```

```python
df=pd.read_csv("employee_data.csv")
df
```

```python
print(df.columns)
```

```python
df.describe()
```

```python
df.notnull()
```

```python
#How many total employees are there?
total_employees=len(df)
print(total_employees)
```

Total number of employees in the dataset

```python
#What is the average salary of employees?
average_salary=df['Salary'].mean()
print(average_salary)
```

Average salary helps understand overall compensation level.

```python
#Which employee has the highest salary?
highest_paid = df.loc[df['Salary'].idxmax()]
print(highest_paid)
```

Identifies top-paid employee.

```python
#What is the average salary for each department?
dept_avg=df.groupby("Department")['Salary'].mean()
print(dept_avg)
```

Helps compare departments financially.

```python
#How many employees are there in each department?
dept_emp_count=df['Department'].count()
print(dept_emp_count)
```

```python
#How many employees are there in each department?
dept_count = df['Department'].value_counts()
print(dept_count)
```

```python
#Which city has the highest number of employees?
city_count = df['City'].value_counts()
print(city_count)

top_city = city_count.idxmax()
print("City with highest employees:", top_city)
```

```python
#Find all employees with salary greater than 60000
high_salary=df[df['Salary']>60000]
print(high_salary)
```

Filters high-earning employees.

```python
#Find employees with experience more than 4 years
experience_emp=df[df['Experience']>4]
print(experience_emp)
```

Identifies senior employees.

```python
#Add a new column "Salary_Level" (High, Medium, Low based on salary)
def salary_level(salary):
    if salary > 70000:
        return "High"
    elif salary >= 50000:
        return "Medium"
    else:
        return "Low"

df['Salary_Level'] = df['Salary'].apply(salary_level)
```

```python
df
```

```python
#Get top 3 highest paid employees
top_3 = df.sort_values(by='Salary', ascending=False).head(3)
print(top_3)
```

```python
# Save processed data
df.to_csv("processed_employee_data.csv", index=False)
```
